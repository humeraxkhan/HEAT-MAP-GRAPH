```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
var= np.linspace(1,10,20).reshape(4,5) #reshape is used for making matrix and linspace 1 to 10 is needed with  20 data elements
var
```

```python
sns.heatmap(var,vmin=0,vmax=14,cmap="PuOr",annot=True)
plt.show()
```
